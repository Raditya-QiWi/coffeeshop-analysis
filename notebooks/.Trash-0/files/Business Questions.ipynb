# Julmlah revenue per hari

```python
%%sql

SELECT
    TO_DATE(`Waktu Transaksi`) AS tanggal_transaksi,
    COUNT(`ID_Transaksi`) AS jumlah_transaksi,
    SUM(`Total Harga`) AS total_revenue
FROM
    Transaksi
GROUP BY
    tanggal_transaksi
ORDER BY
    tanggal_transaksi;
```

# Jumlah penjualan perkategori

```python
%%sql

SELECT
    CASE
        WHEN CAST(`ID Product` AS STRING) LIKE '1%' THEN 'Coffee'
        WHEN CAST(`ID Product` AS STRING) LIKE '2%' THEN 'Tea/dairy'
        WHEN CAST(`ID Product` AS STRING) LIKE '3%' THEN 'Snack'
        ELSE 'Lainnya'
    END AS category_name,
    SUM(Quantity) AS number_of_quantity
FROM
    Transaksi
GROUP BY
    category_name;
```

```python
%%sql
    
SELECT
    k.Nama_Kategori AS category_name,
    SUM(t.Quantity) AS number_of_quantity
FROM
    Transaksi t
JOIN
    Kategori k ON SUBSTRING(CAST(t.`ID Product` AS STRING), 1, 1) = CAST(k.ID_Kategori AS STRING)
GROUP BY
    k.Nama_Kategori;
```

# 5 Product paling laku

```python
%%sql

SELECT
    p.`Produk` AS product_name,
    SUM(t.Quantity) AS number_of_quantity,
    SUM(t.`Total Harga`) AS total_price
FROM
    Transaksi t
JOIN
    Product p ON t.`ID Product` = p.`ID_Produk`
GROUP BY
    p.`Produk`
ORDER BY
    number_of_quantity DESC
LIMIT 5;
```

# Revenue per toko

```python
%%sql

SELECT
    TO_DATE(t.`Waktu Transaksi`) AS date_transaction,
    c.Cabang AS store_name,
    COUNT(t.`ID_Transaksi`) AS number_of_transaction,
    SUM(t.`Total Harga`) AS total_revenue
FROM
    Transaksi t
JOIN
    Cabang c ON t.Branch = c.ID_Cabang
GROUP BY
    date_transaction,
    store_name
ORDER BY
    date_transaction,
    store_name;
```

# Kota paling ramai penjualan

```python
%%sql

SELECT
    c.Cabang AS city_name,
    COUNT(t.`ID_Transaksi`) AS total_transaction
FROM
    Transaksi t
JOIN
    Cabang c ON t.Branch = c.ID_Cabang
GROUP BY
    c.Cabang
ORDER BY
    total_transaction DESC
LIMIT 1;
```

```python
%%sql

SELECT
    c.Cabang AS city_name,
    SUM(t.`Quantity`) AS total_transaction
FROM
    Transaksi t
JOIN
    Cabang c ON t.Branch = c.ID_Cabang
GROUP BY
    c.Cabang
ORDER BY
    total_transaction DESC
LIMIT 1;
```

# Pembayaran non-Cash paling banyak

```python
%%sql

SELECT
    c.Cabang AS city_name,
    SUM(t.`Total Harga`) AS Jumlah
FROM
    Transaksi t
JOIN
    Cabang c ON t.Branch = c.ID_Cabang
WHERE
    t.`Jenis Transaksi` <> 'Cash'
GROUP BY
    c.Cabang
ORDER BY
    Jumlah DESC
LIMIT 1;
```
